```python
#Необходимые библиотеки
import csv
import random
from datetime import datetime, timedelta
from tkinter import *
from tkinter import ttk
from matplotlib.figure import Figure 
from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg, NavigationToolbar2Tk) 
import matplotlib.pyplot as plt
```

```python
# Генерация данных для таблицы csv формата
def generate_date():
    start_date = datetime(2020, 9, 21)
    end_date = datetime(2022, 7, 15)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y-%m-%d')

prod = [
    ('Молоко', 'Молочная продукция', random.randint(1, 200), random.randint(8500, 14000)),
    ('Кефир', 'Молочная продукция', random.randint(1, 65), random.randint(40000, 82000)),
    ('Йогурт', 'Молочная продукция', random.randint(1, 100), random.randint(17000, 75000)),
    ('Милка', 'Шоколад', random.randint(1, 2000), random.randint(1200, 2100)),
    ('Сникерс', 'Шоколад', random.randint(1, 1500), random.randint(2500, 4200)),
    ('Твикс', 'Шоколад', random.randint(1, 750), random.randint(5400, 8700)),
    ('Свинина', 'Мясо', random.randint(1, 1000), random.randint(4500, 19000)),
    ('Курица', 'Мясо', random.randint(1, 850), random.randint(200, 1500)),
    ('Говядина', 'Мясо', random.randint(1, 800), random.randint(600, 2000)),
    ('Лаваш', 'Хлеб', random.randint(1, 1700), random.randint(300, 1200)),
    ('Булочка с маком', 'Хлеб', random.randint(1, 1900), random.randint(500, 1800)),
    ('Пирожок с мясом', 'Хлеб', random.randint(1, 1000), random.randint(650, 1250)),
    ('Святой источник', 'Напитки', random.randint(1, 500), random.randint(3500, 8000)),
    ('Байкал', 'Напитки', random.randint(1, 600), random.randint(1800, 3000)),
    ('Витель', 'Напитки', random.randint(1, 250), random.randint(6000, 9500))
]

with open('Shop.csv', 'w',  newline='') as file:
    writer = csv.writer(file, delimiter='|')

    # Записываем заголовок
    writer.writerow(['Номер заказа', 'Дата заказа', 'Название товара', 'Категория товара', 'Количество продаж', 'Цена за единицу', 'Общая стоимость'])

    ob_vir = 0
    
    # Записываем данные
    for i in range(1, 100):
        random_date = generate_date()
        
        random_product = random.choice(prod)
        product_name = random_product[0]
        category = random_product[1]
        amount = random_product[2]
        price_product = random_product[3]
        
        ob_vir += amount*price_product
        
        writer.writerow([random.randint(1, 100000000), random_date, product_name, category, amount, price_product, amount*price_product])
        
```
